```python
from tpot import TPOTClassifier, TPOTRegressor 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np
import time
import pandas as pd
```

# Standard Required Code

```python
#Drop missing values where percent_missing >= 33%
#see which columns have missing values, sort according to percent missing
df = pd.read_csv('/content/drive/MyDrive/Test_Data_Scientist/Sample_Data_03/dataset_03_with_header.csv')
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})

missing_value_df.sort_values('percent_missing', inplace=True)

missing_value_df_only = missing_value_df[missing_value_df['percent_missing'] != 0]

missing_value_df_to_drop = missing_value_df[missing_value_df['percent_missing'] >= 33]
# print("missing_value_df_to_drop:",missing_value_df_to_drop)
col_to_drop = missing_value_df_to_drop['column_name'].to_list()
# print("col_to_drop:",col_to_drop)

df_dropped_33perc = df.copy()
df_dropped_33perc.drop(col_to_drop, axis=1, inplace = True)
df_dropped_33perc

train_size = int(len(df_dropped_33perc)*0.9)
df_train = df_dropped_33perc[:train_size]
df_test = df_dropped_33perc[train_size:]

# train_path = './dataset/Sample_Data_03/dataset_03_with_header_dropped_33perc_train.csv'
# test_path = './dataset/Sample_Data_03/dataset_03_with_header_dropped_33perc_test.csv'

# df_train.to_csv(train_path)
# df_test.to_csv(test_path)
print("missing_value_df_to_drop: (to drop the following columns)")
missing_value_df_to_drop
```
